```python
#Switch to phoneme dataframe
pcu_matrix = []
for idx, row in groep4DF_kz.iterrows():
    prompt = row["target"]
    pcus_target_graph = row["pcus_target_graph"]
    pcus_target_phon = row["pcus_target_phon"]
    pcus_orig_graph = row["pcus_orig_graph"]

    if(len(pcus_target_graph) == len(pcus_target_phon) and len(pcus_target_graph) == len(pcus_orig_graph)):
        for idx, item in enumerate(pcus_target_graph):
            pcu_matrix.append([pcus_target_graph[idx], pcus_target_phon[idx], pcus_orig_graph[idx], prompt,  pcus_target_graph[idx] == pcus_orig_graph[idx]])
    else:
        print(pcus_target_graph, pcus_target_phon, pcus_orig_graph)

groep4DF_kz_pcu_df = pd.DataFrame(pcu_matrix, columns = ["pcu", "target_phon", "orig_pcu", "prompt", "correct"])
groep4DF_kz_pcu_df.head(2)
```

```python
unique_inc_pcus = list(groep4DF_kz_pcu_df[groep4DF_kz_pcu_df['correct'] == False]["pcu"].value_counts().index)
unique_inc_pcus

outputMatrix = []
for unique_pcu in unique_inc_pcus:
    df_selection = groep4DF_kz_pcu_df[groep4DF_kz_pcu_df['pcu'] == unique_pcu]

    # How often the unique_pcu is written
    total_length = len(df_selection)

    # How often the unique_pcu is written correct
    correct_count = len(df_selection[df_selection['correct'] == True])

    # How often the unique_pcu is written incorrect
    incorrect_count = len(df_selection[df_selection['correct'] == False])

    # In which prompts the PCU occurs
    prompt_occurence = ", ".join(list(set(list(df_selection['prompt']))))

    # List of PCUs with which target PCU is substituted
    unique_orig_graphs = df_selection['orig_pcu'].value_counts().index

    orig_pcu_string_list = []
    for unique_orig_graph in unique_orig_graphs:
        orig_pcu_string_list.append(unique_orig_graph + " (" + str(round((len(df_selection[df_selection['orig_pcu'] == unique_orig_graph])/total_length)*100, 2)) + ");")

    outputMatrix.append([unique_pcu, prompt_occurence, total_length, incorrect_count, round((incorrect_count/total_length)*100,2), len(orig_pcu_string_list)," ".join(orig_pcu_string_list)])

errorTableFile = os.path.join(lrecDataDir, '6-error-table-basiscript-g4p1.tsv')
errorTable = pd.DataFrame(outputMatrix, columns = ['pcu', 'prompts', 'count', 'inc_abs', 'inc_rel', 'count', 'substitutedby'])
errorTable["latex_seperator"] = ["\\\\"] * len(errorTable)
errorTable.to_csv(errorTableFile, sep='&', index=False)
```

```python
# Wat voor soort fouten zijn er gemaakt?
def klankzuiverePCUreadIncorrect(row):
    target_phones = row['target_phones']
    target_pcus = row['pcus_target_graph']
    original_pcus = row['pcus_orig_graph']
    klankzuiverheid_pcus = row['klankzuiver']

    if len(target_phones) == len(target_pcus) and len(target_phones) == len(original_pcus):
        return [target_pcus[idx] + "_" + target_phones[idx] + "_" + original_pcus[idx] + "_" + "".join(target_pcus) + "_" + "".join(original_pcus) for idx, phone in enumerate(target_pcus) if (target_pcus[idx] != original_pcus[idx]) and (klankzuiverheid_pcus[idx] == True)]

    else:
        print(row["original"], target_phones, target_pcus, original_pcus, klankzuiverheid_pcus)

klankzuiverTargetDF.loc[:,'klankzuiver_pcus_read_incorrect'] = klankzuiverTargetDF.apply(klankzuiverePCUreadIncorrect, axis = 1)
```
